```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?

# ans 1:-

'''

Flask is a lightweight and flexible web framework for Python.
Its advantages include simplicity, flexibility, minimal boilerplate code,
extensibility through third-party extensions, and a large community and ecosystem.
It's great for beginners and experienced developers alike,
offering the tools needed to build anything from small prototypes to large-scale applications
efficiently.

Advantages:-

    Simplicity: Flask is known for its simplicity and minimalistic design,
    making it easy to learn and use, especially for beginners.

    Flexibility: It provides developers with the freedom to structure their
    applications according to their specific needs and preferences,
    allowing for easy integration of different libraries and tools.

    Minimal Boilerplate: Flask minimizes the amount of boilerplate code required,
    resulting in faster development and cleaner, more concise code.

    Extensibility: Flask's architecture allows for easy integration of third-party
    extensions, providing additional functionalities such as authentication,
    database integration, and form validation.

    Large Community and Ecosystem: Flask has a large and active community of
    developers who contribute to its development, provide support,
    and create a wide range of third-party libraries and tools.
    This ensures ongoing support and a wealth of resources for developers.
'''
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

# ans 3:-

'''
App routing in Flask refers to the process of mapping URLs to specific functions in your application.
It's a fundamental aspect of building web applications with Flask.

In Flask, we use the @app.route() decorator to define routes.
This decorator tells Flask which function should be called when
a particular URL is requested by a client.

reasons for routers:-
Handling URLs: App routes define how the application responds to different URLs.

Organizing Code: Routes help organize functionality into distinct functions,
making the codebase modular.

RESTful APIs: Routes are essential for building RESTful APIs,
with each route corresponding to a specific endpoint.

Dynamic Content: Routes enable the generation of dynamic content
based on user input or application state.
'''
```
